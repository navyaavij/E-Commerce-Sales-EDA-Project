US E-commerce transaction data containing customer details, product info, device usage, delivery type, and transaction outcomes.

Problem statement : Analyzing customer purchasing behavior and segment customers to identify high-value users and sales trends.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Us-Ecommerce Dataset-checkpoint.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#checking_missing_values
df.isnull().sum() 
```

```python
#filling_missing_values
df.fillna("Unkown", inplace = True)
df.isnull().sum()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df.columns.tolist()
```

```python
#Rename columns (remove spaces)
df.columns = df.columns.str.strip()
```

```python
df.columns.tolist()
```

```python

#Convert: Date → datetime & Time → datetime or hour

df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y', errors='coerce')
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce')
df['Year_Month'] = pd.to_datetime(df['Year_Month'], format='%Y-%m', errors='coerce')

```

```python
#Convert : Amount US$, Individual_Price_US$ → numeric


df['Amount US$'] = (df['Amount US$'].replace(r'[\$,]', '', regex=True).replace('Unkown', pd.NA))
df['Amount US$'] = pd.to_numeric(df['Amount US$'], errors='coerce')
df['Individual_Price_US$'] = (df['Individual_Price_US$'].replace(r'[\$,]','',regex=True).replace('Unkown', pd.NA))
df['Individual_Price_US$'] = pd.to_numeric(df['Individual_Price_US$'], errors = 'coerce')
```

```python
delivery_failure = df.groupby('Delivery_Type')['is_success'].mean()
delivery_failure = (1 - delivery_failure) * 100
delivery_failure.plot(kind='bar')
plt.title('Failure Rate by Delivery Type')
plt.ylabel('Failure %')
plt.show()
```

KEY INSIGHTS
1.Member customers drive the business
   - Member users account for over 95% of total transactions, clearly outperforming guests and new users.
   - This indicates strong customer loyalty and highlights the success of the membership model.
2.High-value customers form a significant segment
    - A large majority of customers fall under the high-value category, contributing disproportionately to total revenue.
    - Retention and personalized offers for these customers can significantly boost profitability.
3. Transaction amounts are highly right-skewed
    - Most transactions occur at lower values, while a small number of extremely high-value transactions create a long right tail.
    - This suggests the presence of premium purchases or bulk buyers.
4. Peak transaction activity occurs at specific hours
    - Hourly analysis reveals clear peaks during certain hours of the day, indicating time-based buying behavior.
    - Marketing campaigns and offers can be optimized around these peak hours.
5. Zero and failed transactions are non-negligible
    - A noticeable number of transactions have zero or failed amounts, pointing toward payment failures, cancellations, or system issues.
    - Improving payment reliability could directly improve revenue realization.
6. Guest and new users contribute minimally
    - Guest and first-time users together form less than 5% of total activity, with very low revenue contribution.
    - This suggests an opportunity to improve onboarding and first-time user experience.
7. Revenue concentration indicates optimization potential
    - Revenue is concentrated among specific users and patterns rather than evenly distributed.
    - Targeted strategies can yield better ROI than broad, untargeted campaigns.

This analysis uncovers strong customer loyalty, revenue concentration among high-value members, and clear behavioral patterns across time, offering actionable insights to optimize retention, revenue, and operational efficiency.
